### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
!pip install -q pandas
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable
%load_ext sql
import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
%sql sqlite:///my_data1.db
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null

# Get the unique launch sites
launch_sites = %sql SELECT DISTINCT Landing_Outcome FROM SPACEXTABLE

# Display the launch sites
print(launch_sites)

#Or, to print in a more user-friendly way:

launch_sites_list = [row[0] for row in launch_sites] #Extract the names from the SQL result
print("\nUnique Launch Sites:")
for site in launch_sites_list:
    print(site)

# Or, using Pandas (after retrieving the data):
launch_sites_df = pd.read_sql_query("SELECT DISTINCT Landing_Outcome FROM SPACEXTABLE", con)
print("\nUnique Launch Sites (Pandas):\n", launch_sites_df)

launch_sites_list_pandas = launch_sites_df['Landing_Outcome'].tolist()
print("\nUnique Launch Sites (Pandas List):")
for site in launch_sites_list_pandas:
    print(site)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
!pip install -q pandas
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable
%load_ext sql
import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
%sql sqlite:///my_data1.db
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null

# SQL Query to select 5 records where LaunchSite begins with 'CCA'
cca_launches = %sql SELECT * FROM SPACEXTABLE WHERE Launch_Site LIKE 'CCA%' LIMIT 5;

# Display the results
print(cca_launches)

#If you want to print the results in a more formatted way:
cca_launches_list = [row for row in cca_launches] #Extract the rows from the SQL result

unique_launch_pads_df = pd.read_sql_query("SELECT DISTINCT Launch_Pad FROM SPACEXTABLE", con)

print(unique_launch_pads_df)  # Print the DataFrame

# To get the unique values as a list:
unique_launch_pads_list = unique_launch_pads_df['Launch_Pad'].tolist()
print(unique_launch_pads_list)

# Or, to print each launch pad on a separate line:
for launch_pad in unique_launch_pads_list:
    print(launch_pad)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
!pip install -q pandas
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable
%load_ext sql
import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
%sql sqlite:///my_data1.db
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
# Filter for NASA (CRS) missions
nasa_crs_missions = df[df['Customer'] == 'NASA (CRS)'].copy() # create a explicit copy to avoid the warning

# Convert 'PAYLOAD_MASS__KG_' to numeric, handling errors (important!)
nasa_crs_missions.loc[:, 'PAYLOAD_MASS__KG_'] = pd.to_numeric(nasa_crs_missions['PAYLOAD_MASS__KG_'], errors='coerce') # use .loc to avoid the warning

# Calculate the total payload mass, ignoring NaN values
total_payload_mass = nasa_crs_missions['PAYLOAD_MASS__KG_'].sum(skipna=True) #skipna=True ignores NaN values

# Print the result
print(f"Total payload mass carried by NASA (CRS) boosters: {total_payload_mass} kg")

# To handle potential missing or non-numeric values in the PAYLOAD_MASS__KG_ column
# It's good practice to check for NaN values after the conversion:
nan_count = nasa_crs_missions['PAYLOAD_MASS__KG_'].isna().sum()
if nan_count > 0:
  print(f"Warning: {nan_count} payload mass values were missing or invalid and were treated as 0 in the calculation.")
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
!pip install -q pandas
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable
%load_ext sql
import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
%sql sqlite:///my_data1.db
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null

# Filter for booster version F9 v1.1
f9_v1_1_missions = df[df['Booster_Version'] == 'F9 v1.1'].copy() # create explicit copy to avoid warning

# Convert 'PAYLOAD_MASS__KG_' to numeric, handling errors
f9_v1_1_missions.loc[:, 'PAYLOAD_MASS__KG_'] = pd.to_numeric(f9_v1_1_missions['PAYLOAD_MASS__KG_'], errors='coerce')

# Calculate the average payload mass, ignoring NaN values
average_payload_mass = f9_v1_1_missions['PAYLOAD_MASS__KG_'].mean(skipna=True)

# Print the result
print(f"Average payload mass carried by F9 v1.1 boosters: {average_payload_mass} kg")

# Handle potential missing or non-numeric values
nan_count = f9_v1_1_missions['PAYLOAD_MASS__KG_'].isna().sum()
if nan_count > 0:
    print(f"Warning: {nan_count} payload mass values were missing or invalid and were treated as NaN in the calculation.")
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
!pip install -q pandas
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable
%load_ext sql
import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
%sql sqlite:///my_data1.db
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null

# Filter for successful ground pad landings
successful_ground_pad_landings = df[df['Landing_Outcome'] == 'Success (ground pad)'].copy() #add .copy()

# Convert 'Date' to datetime objects for proper sorting
successful_ground_pad_landings.loc[:, 'Date'] = pd.to_datetime(successful_ground_pad_landings['Date']) #use .loc

# Find the earliest (minimum) date
first_successful_landing_date = successful_ground_pad_landings['Date'].min()

# Print the result
print(f"The date of the first successful landing outcome on a ground pad was: {first_successful_landing_date.date()}")
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
!pip install -q pandas
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable
%load_ext sql
import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
%sql sqlite:///my_data1.db
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null

# Filter for successful drone ship landings and payload mass criteria
filtered_boosters = df[
    (df['Landing_Outcome'] == 'Success (drone ship)') &
    (df['PAYLOAD_MASS__KG_'] > 4000) &
    (df['PAYLOAD_MASS__KG_'] < 6000)
]

# Get the list of booster names
booster_names = filtered_boosters['Booster_Version'].tolist()

# Print the result
print("Boosters with successful drone ship landings and payload mass between 4000 and 6000 kg:")
for name in booster_names:
    print(name)
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
!pip install -q pandas
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable
%load_ext sql
import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
%sql sqlite:///my_data1.db
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null

# Find the maximum payload mass
max_payload = df['PAYLOAD_MASS__KG_'].max()

# Use a subquery to filter booster versions with the maximum payload mass
max_payload_boosters = df[df['PAYLOAD_MASS__KG_'] == max_payload]['Booster_Version'].tolist()

# Print the result
print("Booster versions with maximum payload mass:")
for booster in max_payload_boosters:
    print(booster)
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
!pip install -q pandas
!pip install sqlalchemy==1.3.9
!pip install ipython-sql
!pip install ipython-sql prettytable
%load_ext sql
import csv, sqlite3
import prettytable
prettytable.DEFAULT = 'DEFAULT'

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
%sql sqlite:///my_data1.db
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null

# Filter for the specified criteria
filtered_records = df[
    (df['Landing_Outcome'] == 'Failure (drone ship)') &
    (df['Date'].str.startswith('2015'))
].copy() #add .copy()

# Extract month and year
filtered_records.loc[:, 'Month'] = filtered_records['Date'].str[5:7] #use .loc

# Select and print the desired columns
result = filtered_records[['Month', 'Landing_Outcome', 'Booster_Version', 'Launch_Site']]

# Convert month number to month name.
month_dict = {
    '01': 'January', '02': 'February', '03': 'March', '04': 'April',
    '05': 'May', '06': 'June', '07': 'July', '08': 'August',
    '09': 'September', '10': 'October', '11': 'November', '12': 'December'
}

result.loc[:, 'Month'] = result['Month'].map(month_dict) #use .loc

print(result)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
import pandas as pd

# Sample DataFrame (replace with your actual data)
df = pd.DataFrame({
    'Date': ['2010-01-01', '2011-02-15', '2015-03-20', '2016-06-04', '2017-03-21', '2010-06-05'],
    'Landing_Outcome': ['Success (ground pad)', 'Failure (drone ship)', 'Success (ground pad)', 
                       'Failure (drone ship)', 'Success (ground pad)', 'Success (ground pad)']
})

# Convert 'Date' to datetime objects
df['Date'] = pd.to_datetime(df['Date'], utc=True)

# Filter data between the specified dates
filtered_df = df[(df['Date'] >= '2010-06-04') & (df['Date'] <= '2017-03-20')]

# Count occurrences of each landing outcome
outcome_counts = filtered_df['Landing_Outcome'].value_counts()

# Rank the outcomes by count in descending order
ranked_outcomes = outcome_counts.rank(method='dense', ascending=False)

# Create a DataFrame with the ranked outcomes
ranked_df = pd.DataFrame({'Landing_Outcome': ranked_outcomes.index, 'Rank': ranked_outcomes})

print(ranked_df)
```
