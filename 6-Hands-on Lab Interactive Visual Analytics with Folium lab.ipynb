Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
import piplite
await piplite.install(['folium'])
await piplite.install(['pandas'])
import folium
import pandas as pd
# Import folium MarkerCluster plugin
from folium.plugins import MarkerCluster
# Import folium MousePosition plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(dataset_part_2_csv)
#spacex_df.tail(10)

# Start location 
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

marker_cluster = MarkerCluster().add_to(site_map)

# Group by Launch Site and calculate success rate
launch_site_success = spacex_df.groupby('Launch Site')['class'].mean().reset_index()

# Merge success rate back into launch_sites_df
launch_sites_df = pd.merge(launch_sites_df, launch_site_success, on='Launch Site', how='left')

# Iterate through launch sites and add circles and markers
for index, row in launch_sites_df.iterrows():
    site_name = row['Launch Site']
    lat = row['Lat']
    long = row['Long']
    success_rate = row['class']  # Access the calculated success rate
    coordinate = [lat, long]

    # Determine circle color based on success rate
    if success_rate >= 0.8:  # Example threshold (adjust as needed)
        color = '#008000'  # Green for high success rate
    elif success_rate >= 0.5:
        color = '#FFA500'  # Orange for medium success rate
    else:
        color = '#FF0000'  # Red for low success rate

    # Add a circle for each launch site, colored by success rate
    circle = folium.Circle(
        coordinate,
        radius=1000,
        color=color,  # Use the determined color
        fill=True
    ).add_child(folium.Popup(f"{site_name}: Success Rate = {success_rate:.2f}"))  # Popup shows success rate
    site_map.add_child(circle)

    # Add marker (optional - you can keep or remove it)
    marker = folium.map.Marker(
        coordinate,
        icon=folium.DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{site_name}</b></div>'
        )
    )
    marker.add_to(marker_cluster)
```
